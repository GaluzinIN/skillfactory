```python
answers = {} # создадим словарь для ответов

# тут другие ваши предобработки колонок например:

#the time given in the dataset is in string format.
#So we need to change this in datetime format

#  0.imdb_id - Ид фильма
#  1.budget - бюджет фильма
#  2.revenue - сборы
#  3.original_title - название фильма
#  4.cast - в ролях
#  5.director - режиссер (директор картины)
#  6.tagline - тэг
#  7.overview - краткий обзор фильма
#  8.runtime - длина фильма в минутах
#  9.genres - жанр фильма
# 10.production_companies - киностудия создатель
# 11.release_date - дата выхода на экран фильма
# 12.vote_average - рейтинг фильма
# 13.release_year - год выхода фильма

# Расчет прибыли 
data['profit'] = data['revenue'] - data['budget']

# Здесь храним количество фильмов по конкретным жанрам. Нужен для вопроса №11.
data_genres = pd.Series([0, 0, 0, 0, 0],
                 index = ['Action', "Adventure", "Drama", "Comedy", "Thriller"])

# Функция заменяет дату релиза месяцем релизадля вопросов №21-22
def up_mon(x):
    return (x[:x.find('/')])

# Функция выдает 1 если месяц зимний, для вопроса №23
def winter_comm(x):
    if int(x[:x.find('/')]) < 3:
        return 1
    elif int(x[:x.find('/')]) == 12:
        return 1
    else:
        return 0
```

# 27. Какие актеры чаще всего снимаются в одном фильме вместе?


ВАРИАНТ 2

```python
answers['27'] = 'Daniel Radcliffe & Rupert Grint'

# Разбиваем строку на части
data27 = data['cast'].str.split('|')
# Создаем пустой для того чтобы при перезапуске не происходило накапливания данных
resultat = list()
# бежим по строкам я знаю что их 1889 и создаем список пар артистов снимающихся в одном фильме
for i in range(1889):
    resultat += list(combinations(data27[i], 2))
# преобразуем список в серию для того чтобы воспользоваться функцией value_counts()
resultat = pd.Series(resultat)
resultat.value_counts()
```
